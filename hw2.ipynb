```python
import numpy as np
import pandas as pd
import mnist
import keras

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
df_train = pd.read_csv('./fashionmnist/fashion-mnist_train.csv')
df_test = pd.read_csv('./fashionmnist/fashion-mnist_test.csv')
```

```python
df_train.head()
```

```python
df_label_train = df_train['label']
df_label_test = df_test['label']
```

```python
df_label_train.unique()
```

```python
df_img_train = df_train.drop('label', axis=1)
df_img_test = df_test.drop('label', axis=1)
```

```python
df_train.info()
```

```python
df_img_train = (df_img_train/255) * 2 - 1
df_img_test = (df_img_test/255) * 2 - 1
```

```python
df_img_train.head()
```

```python
df_img_train.describe()
```

```python
print(df_img_train.shape)
print(df_img_test.shape)
```

```python
model = Sequential([
  Dense(64, activation='relu', input_shape=(784,)),
  Dense(64, activation='relu'),
  Dense(10, activation='softmax'),
])
```

```python
model.compile(
  optimizer='adam',
  loss='categorical_crossentropy',
  metrics=['accuracy'],
)
```

```python
model.fit(
  df_img_train,
  to_categorical(df_label_train),
  epochs=10,
  batch_size=32,
)
```

```python
model.evaluate(
  df_img_test,
  to_categorical(df_label_test)
)
```

```python
model.save_weights('model.h10')
```

```python
predictions = model.predict(df_img_test)
```
